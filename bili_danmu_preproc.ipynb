```python
import os
import json

file_name = 'response_zzz1.4开服测试_20241216_163827.txt'
output_name = 'processed_zzz1.4开服测试_20241216_163827.txt'

def process_danmu(file_path):
    # 读取文件内容
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    processed_data = []

    for line in lines:
        try:
            # 将每一行的字符串解析为字典
            danmu = json.loads(line.strip())
            
            text = danmu.get('text', "")
            print(text)
            user = danmu.get('nickname', "")
            uid = danmu.get('uid', "")
            avatar = danmu.get('user', {}).get('base', {}).get('face', "")

            # 提取粉丝牌信息
            medal_info = danmu.get('medal', [])
            medal = medal_info[1] if len(medal_info) > 1 else ""
            medal_level = medal_info[0] if len(medal_info) > 0 else ""

            # 时间和日期
            timeline = danmu.get('timeline', "")
            date, time_part = timeline.split(' ') if ' ' in timeline else ("", "")

            processed_data.append({
                "date": date,
                "time": time_part,
                "text": text,
                "user": user,
                "uid": uid,
                "avatar": avatar,
                "medal": medal,
                "medal_level": medal_level
            })
        except json.JSONDecodeError as e:
            print(f"解析JSON出错: {e}")
        except Exception as e:
            print(f"处理弹幕时出错: {e}")
            
    return processed_data



def main():
    input_dir = "live_raw_response"
    output_dir = "live_pre_processing"
    os.makedirs(output_dir, exist_ok=True)

    # 用户输入文件名
    input_file = os.path.join(input_dir, file_name)

    if not os.path.exists(input_file):
        print(f"文件 {input_file} 不存在！")
        return

    output_name = file_name
    
    output_file = os.path.join(output_dir, output_name)

    processed_data = process_danmu(input_file)
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(processed_data, f, ensure_ascii=False, indent=4)
    print(f"已保存处理后的文件到: {output_file}")
    
if __name__ == "__main__":
    main()
```
